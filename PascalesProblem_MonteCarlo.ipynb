```python
import random
```

```python
def rollDie():
  return random.choice([1,2,3,4,5,6])
```

Probability that double 6's will be rolled

```python
def checkProb(numTrials, numRolls):
  # numTrials > 0
  # Track number of double 6's rolled
  numWins = 0.0
  for i in range(numTrials):
    for j in range(numRolls):
      # die 1
      d1 = rollDie()
      # die 2
      d2 = rollDie()
      if d1 == 6 and d2 == 6:
        numWins += 1
        break
    print("Probability of winning after {} trials is {} ".format(i, (numWins/numTrials)))
```

<h1>Pascale's Problem</h1>
Within 24 rolls of a pair of dice, will a double six be rolled?

```python
# 1000000 trials of 24 rolls each
checkProb(1000000,24)
```
